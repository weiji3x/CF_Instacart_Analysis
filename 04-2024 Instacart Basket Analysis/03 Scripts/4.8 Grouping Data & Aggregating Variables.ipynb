### 4.6 The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
- If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.
- If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

```python
# Create a new 'avg_spending' column 
# and apply transform() function on 'prices' column to generate the average spending for each user
ords_prods_merge['avg_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(20)
```

```python
# If cusomter have an average spending of < 10, assign 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_spending'] < 10, 'Spender_status'] = 'Low spender'
```

```python
# If cusomter have an average spending of >= 10, assign 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_spending'] >= 10, 'Spender_status'] = 'High spender'
```

```python
ords_prods_merge.head(20)
```

```python
# Check count for spender status
ords_prods_merge['Spender_status'].value_counts(dropna = False)
```

### 4.7 In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column. The criteria for the flag should be as follows:
- If the median of days_since_prior_order is higher than 20, then the customer should be labeled a Non-frequent customer.
- If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a Regular customer.
- If the median is lower than or equal to 10, then the customer should be labeled a Frequent customer.

```python
# Create a new 'median_order_days' column 
# and apply transform() function on 'prices' column to generate the average spending for each user
ords_prods_merge['median_order_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the 'median_order_days' column is created
ords_prods_merge.head()
```

```python
# If cusomter have median_order_days > 20, assign 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_order_days'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
# If cusomter have median_order_days > 10 and <= 20, assign 'Regular customer'
ords_prods_merge.loc[(ords_prods_merge['median_order_days'] > 10) & (ords_prods_merge['median_order_days'] <= 20), 'order_frequency'] = 'Regular customer'
```

```python
# If cusomter have median_order_days > 20, assign 'Frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_order_days'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Check value count for 'order_frequency' column
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```

```python
df_new2 = ords_prods_merge.loc[ords_prods_merge['order_frequency'] == 'nan']
```

```python
df_new2
```

#### Seems like the order_frequency with nan belongs to a customer who only ordered once, thus there was no days_since_prior_order.

```python
# Setting the order_frequency with 'nan' to 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['order_frequency'] == 'nan', 'order_frequency'] = 'Non-frequent customer'
```

```python
# Check value count for 'order_frequency' column
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```
